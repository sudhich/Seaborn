```python
sns.relplot(x='tip', y='total_bill',data=tips)
#Relation bitween tip and total_bill
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, hue='smoker')
#Relation between tip and total_bill who is smoker and nonsmoker(groupby).
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, hue='day')
#Relation between tip and total_bill based on day. suppose in restrorant red representation sunday.
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, hue='smoker')
#Relation between tip and total_bill who is smoker and nonsmoker(groupby).
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, hue='day', style='smoker')
#Suppose in sunday how many moker and non smoker.
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, hue='day', style='smoker',size='size')
#Suppose in sunday how many moker and non smoker people reached in group at restorant.
```

```python
sns.relplot(x='tip', y='total_bill',data=tips, size='tip')
#who gives more tips according to size.
```
